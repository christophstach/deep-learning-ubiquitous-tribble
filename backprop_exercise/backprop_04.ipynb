## Lösung

```python
f = Digraph('finite_state_machine', filename='forward.gv')
f.attr(label=r'\n\n forward')
f.attr(fontsize='20')

f.attr(rankdir='LR', size='8,5')

f.attr('node', shape='circle')
f.node('1')
f.node('2')
f.node('3')
f.node('4')

f.edge('1','+ ')
f.edge('2','+ ')

f.edge('+ ', '*2',label='3')
f.edge('*2','()²',label='6')
f.edge('()²','max',label='36')

f.edge('3', '+')
f.edge('4', '+')
f.edge('+','x/2',label='7')
f.edge('x/2','()³',label='3,5')
f.edge('()³','max',label='42.875')



f.edge('max','out',label='42.875')
f.node('out', shape='Msquare')


f
```

```python
f = Digraph('finite_state_machine', filename='forward.gv')
f.attr(label=r'\n\n backwards')
f.attr(fontsize='20')

f.attr(rankdir='LR', size='8,5')

f.attr('node', shape='circle')
f.node('1')
f.node('2')
f.node('3')
f.node('4')

f.edge('1','+ ')
f.edge('2','+ ')

f.edge('+ ', '*2',label='0')
f.edge('*2','()²',label='0')
f.edge('()²','max',label='0')

f.edge('3', '+',label='18,375')
f.edge('4', '+',label='18,375')
f.edge('+','x/2',label='18,375')
f.edge('x/2','()³',label='36,75')
f.edge('()³','max',label='1')



f.edge('max','out',label='1')
f.node('out', shape='Msquare')


f
```

Es fällt auf das der Plus-Operator direkt am Beginn der Funktion nur dafür sorgt, dass $\frac{\partial out}{\partial c}$=$\frac{\partial out}{\partial d}$.

Weiter wird klar, dass der max Operator den zu  berechnenden Aufwand reduziert und nicht erweitert.

$f(a,b,c,d) = max \left\{[(a+b)*2]^{2} , (\frac{c+d}{2})^{3}\right\}$

Damit a bzw. b Auswirkungen auf den Output haben muss die obere Teilfunktion größer werden als die untere.
